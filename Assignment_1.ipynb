```python
import numpy as np
import matplotlib.pyplot as plt
data = np.array([2.3, 2.5, 3.6, 2.8, 3.1, 2.9, 3.2, 2.7, 2.8, 3.0])
num_bins = 5
data_range = np.max(data) - np.min(data)
bin_width = data_range / num_bins
bin_edges = np.linspace(np.min(data), np.max(data), num_bins + 1)
counts, _ = np.histogram(data, bins=bin_edges)
total_points = len(data)
density = counts / (total_points * bin_width)
for i in range(num_bins):
    print(f"Bin {i + 1} ({bin_edges[i]:.2f} to {bin_edges[i + 1]:.2f}): Count = {counts[i]}, Density = {density[i]:.2f}")
plt.hist(data, bins=bin_edges, density=True, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram with Density')
plt.show()
```
